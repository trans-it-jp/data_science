# サンプルデータ

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## scit-lean toy datasets

### あやめ品種データ

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris.keys()
```

```python
for key in iris.keys():
    print(key, ':', iris[key])
```

```python
classes = iris['target_names']
iris_df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_df.head()
```

```python
iris_df['species'] = pd.Categorical.from_codes(iris.target, classes)
iris_df.head()
```

|パラメータ|説明|
|-|-|
|sepal length|がくの長さ|
|sepal width|がくの幅|
|petal length|花びらの長さ|
|petal width|花びらの幅|
|species|種名|

```python
iris_df.shape
```

### 糖尿病データ

```python
from sklearn.datasets import load_diabetes
```

```python
datasets = load_diabetes()
datasets.keys()
```

```python
for key in datasets.keys():
    print(key, ':', datasets[key])
```

```python
diabetes_df = pd.DataFrame(datasets['data'], columns=datasets['feature_names'])
diabetes_df.head()
```

```python
diabetes_df['target'] = datasets['target']
diabetes_df.head()
```

|パラメーター|説明|
|---|---|
|age|年齢|
|sex|性別|
|bmi|BMI|
|bp|平均血圧|
|s1|s1 tc, 血清コレステロール|
|s2|s2 ldl, 低比重リポタンパク質|
|s3|s3 hdl, 高比重リポタンパク質|
|s4|s4 tch, 総コレステロール/HDL|
|s5|s5 ltg, 血清トリグリセリド値の対数|
|s6|s6 glu, 血糖値|
|target|1年後の疾患進行度|

説明変数(target以外）は標準化(平均0 標準偏差1 の分布に加工)されている。

```python
diabetes_df.shape
```

### 手書き数字データ

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
digits.keys()
```

```python
for key in digits.keys():
    print(key, ':', digits[key])
```

```python
digits_df = pd.DataFrame(digits['data'])
digits_df.head()
```

```python
digits_df['target'] = digits['target']
digits_df.head()
```

```python
# 64個の特徴量を8x8の画像として表示
plt.figure(figsize=(8, 8))
for i in range(64):
    plt.subplot(8, 8, i + 1)
    plt.imshow(digits.images[i], cmap='gray')
    plt.axis('off')
```

|パラメーター|説明|
|-|-|
|0〜63|各ピクセルの特徴量(色の濃さ: 0以上16以下の数値)|
|target|記述している数字|
